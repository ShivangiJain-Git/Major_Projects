# Supervised Learning - Bank Loan Prediction

## Context

This case is about a bank (Thera Bank) whose management wants to explore ways of converting its liability customers to personal loan customers (while retaining them as depositors). A campaign that the bank ran last year for liability
customers showed a healthy conversion rate of over 9% success. This has encouraged the retail marketing department to devise campaigns with better target marketing to increase the success ratio with minimal budget.

## Objective

The classification goal is to predict the likelihood of a liability customer buying personal loans.

## Data Description

The file Bank.xls contains data on 5000 customers. The data include customer demographic information (age, income, etc.), the customer's relationship with
the bank (mortgage, securities account, etc.), and the customer response to the last personal loan campaign (Personal Loan). Among these 5000 customers,
only 480 (= 9.6%) accepted the personal loan that was offered to them in the earlier campaign. 

## Attribute Information

1. ID : Customer ID
2. Age : Customer's age in completed years
3. Experience : #years of professional experience
4. Income : Annual income of the customer in dollar
5. ZIP Code : Home Address ZIP code.
6. Family : Family size of the customer
7. CCAvg : Avg. spending on credit cards per month in dollar
8. Education : Education Level.

	1.Undergrad
	2. Graduate
	3. Advanced/Professional
9. Mortgage : Value of house mortgage if any in dollar
10. Personal Loan : Did this customer accept the personal loan offered in the last campaign?
11. Securities Account : Does the customer have a securities account with the bank?
12. CD Account : Does the customer have a certificate of deposit (CD) account with the bank?
13. Online : Does the customer use internet banking facilities?
14. Credit card : Does the customer use a credit card issued by UniversalBank?

##  Split the data into training and test set in the ratio of 70:30 respectively.

```python
from sklearn.model_selection import train_test_split

#Independent variables or Predictor Feature. Here we are dropping ID, Age along with target variable.
X= df.drop(['ID','Personal Loan','Age'], axis =1)

#Dependent variable or Predicted class
y = df['Personal Loan']

#Split data in training and test set in the ratio of 70:30
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.30, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

##  Using different classification models (Logistic,K-NN and Naive Bayes) to predict the likelihood of a customer buying personal loans.

#### -------------------------------------------------------LOGISTIC REGRESSION MODEL--------------------------------------------------------------------------------

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics 
from sklearn.metrics import accuracy_score, confusion_matrix , recall_score, classification_report, average_precision_score, plot_confusion_matrix


lr = LogisticRegression(max_iter=100)
lr.fit(X_train,y_train)

#Predicting on test set
y_pred_lr_test = lr.predict(X_test)

#Predicting for training set
y_pred_lr_train = lr.predict(X_train)

print('Logistic Regression Test accuracy is : ',accuracy_score(y_test,y_pred_lr_test))
# print('Train accuracy is :',accuracy_score(y_train,y_pred_lr_train))
print('Confusion Matrix : \n',confusion_matrix(y_test,y_pred_lr_test))
```

The test accuracy score of .90 looks impressive but do not forget, it is unreliable as it is a score at model level. Let us look at class level, especially the class 1.

```python
print('Classification report :\n', classification_report(y_test,y_pred_lr_test,labels=[0,1]))
```

The precision and recall for class 1 is low...

Precision: Within a given set of positively-labeled results, the fraction that were 
true positives = tp/(tp + fp) , this has to be calculated for each class i.e. 0 and 1 and should be
high for the class less represented, class 1 in our example

Recall: Given a set of positively-labeled results, the fraction of all positives that were 
retrieved = tp/(tp + fn)

Accuracy: tp + tn / (tp + tn + fp +fn) But this measure can be dominated by larger class. 

```python
#We can also calculate threshold using ROC curve
#AUC ROC curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
# roc curve for logistic regression model with optimal threshold
from numpy import sqrt
from numpy import argmax

logit_roc_auc = roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])
fpr, tpr, thresholds = roc_curve(y_test, lr.predict_proba(X_test)[:,1])
# calculate the g-mean for each threshold
gmeans = sqrt(tpr * (1-fpr))
# locate the index of the largest g-mean
ix = argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--',label='No Skill')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

#### ---------------------------------------------------------------------KNN Model------------------------------------------------------------------------------------------------

```python
from sklearn.neighbors import KNeighborsClassifier

#considering K=3
knn3 = KNeighborsClassifier(n_neighbors = 3)
knn3.fit(X_train,y_train)

#Predict for Test set
y_pred_knn3_test = knn3.predict(X_test)
print("KNN Test Accuracy when K=3 : ",accuracy_score(y_test,y_pred_knn3_test))

# K = 5
knn5 =KNeighborsClassifier(n_neighbors = 5)
knn5.fit(X_train,y_train)
y_pred_knn5_test = knn5.predict(X_test)
print("KNN Test Accuracy when K=5 :",accuracy_score(y_test,y_pred_knn5_test))

knn9 = KNeighborsClassifier(n_neighbors = 9)
knn9.fit(X_train, y_train)
y_pred_knn9_test = knn9.predict(X_test)
print("KNN Test Accuracy when K=9 :",accuracy_score(y_test,y_pred_knn9_test))


#Creating odd list to find best K
mylist = list(range(1,20))
neighbors = list(filter(lambda x : x % 2 != 0,mylist))
print(neighbors)
```

```python
ac_scores = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred_val = knn.predict(X_test)
    scores = accuracy_score(y_test,y_pred_val)
    ac_scores.append(scores)
#Changing to misclassification error
MSE = [1- x for x in ac_scores]
best_k = neighbors[MSE.index(min(MSE))]
print('best K is : ',best_k)


plt.plot(neighbors,MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')

#cal accuracy score when k =13
knn_new = KNeighborsClassifier(n_neighbors = 13)
knn_new.fit(X_train,y_train)

#Predict test
y_pred_13_test = knn_new.predict(X_test)
print('KNN Test Accuracy when K=13',accuracy_score(y_test,y_pred_13_test))

print('Classification Report : \n', classification_report(y_test,y_pred_13_test))
```

Here we can precision and recall (least .09) is very low for class 1.
