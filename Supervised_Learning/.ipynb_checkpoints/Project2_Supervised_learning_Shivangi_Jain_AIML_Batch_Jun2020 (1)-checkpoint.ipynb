## 4. Split the data into training and test set in the ratio of 70:30 respectively (5 marks).**

```python
from sklearn.model_selection import train_test_split
#Independent variables or Predictor Features  --Here we are dropping Age also, as we found it is highly corelated with Experience
X= df.drop(['ID','Personal Loan','Age'], axis =1)

#Dependent variable or Predicted class
y = df['Personal Loan']

#Split data in training and test set in the ratio of 70:30
X_train1 , X_test1 , y_train1 , y_test1 = train_test_split(X,y,test_size=0.30, random_state=1)
```

## 5. Use different classification models (Logistic,K-NN and Naive Bayes) to predict the likelihood of a customer buying personal loans (15 marks).**

#### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics 
from sklearn.metrics import accuracy_score, confusion_matrix , recall_score, classification_report, average_precision_score, plot_confusion_matrix

print('Shape of training data is :',X_train1.shape)   #Checking dimension of training data
print('Shape of testing data is :',X_test1.shape)    #Checking dimension of test data

logreg = LogisticRegression(max_iter=100)
logreg.fit(X_train1,y_train1)

#Predict for X_test1
y_pred1_test = logreg.predict(X_test1)

#Predict for X_train1
y_pred1_train = logreg.predict(X_train1)

print('Test accuracy is : ',accuracy_score(y_test1,y_pred1_test))
print('Train accuracy is :',accuracy_score(y_train1,y_pred1_train))
print('Classification report \n')
print(classification_report(y_test1,y_pred1_test,labels=[1, 0]))
```

Here precision and recall is very low. Lets standardize the data

```python
#Standardizing Data
# import zscore for scaling the data
from scipy.stats import zscore

#apply zscore on independent features
Xtrain_std = X_train1.apply(zscore)
Xtest_std = X_test1.apply(zscore)

#Fit model on scaled data
logreg_std = LogisticRegression(random_state=0)
logreg_std.fit(Xtrain_std,y_train1)

#making prediction on scaled data
y_pred_train_std = logreg_std.predict(Xtrain_std)
y_pred_test_std = logreg_std.predict(Xtest_std)

#Calculating score
print('Scode of std train data',logreg_std.score(Xtrain_std,y_pred_train_std))
print('Score of std test data',logreg_std.score(Xtest_std,y_pred_test_std))

print(classification_report(y_test1,y_pred_test_std,labels=[1, 0]))

```

Now, we can see precision and recall both got improved after scaling data and also Accuracy score is very good.

```python
#We can also calculate threshold using ROC curve
#AUC ROC curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
# roc curve for logistic regression model with optimal threshold
from numpy import sqrt
from numpy import argmax

logit_roc_auc = roc_auc_score(y_test1, logreg.predict_proba(X_test1)[:,1])
fpr, tpr, thresholds = roc_curve(y_test1, logreg.predict_proba(X_test1)[:,1])
# calculate the g-mean for each threshold
gmeans = sqrt(tpr * (1-fpr))
# locate the index of the largest g-mean
ix = argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--',label='No Skill')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
# y_pred_test_std = logreg_std.predict(Xtest_std)
THRESHOLD = 0.077
y_pred_test_std_final = np.where(logreg_std.predict_proba(Xtest_std)[:,1]>THRESHOLD,1,0)
logreg_std.score(Xtest_std,y_pred_test_std_final)


```

But, using this threshold, we observe accuracy has reduced

#### K-NN

```python
from sklearn.neighbors import KNeighborsClassifier

#considering K=3
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train1,y_train1)

#Predict for Test
y_pred = knn.predict(X_test1)
print("KNN Test Accuracy when K=3 : ",accuracy_score(y_test1,y_pred))

#Predict train
y_pred_train = knn.predict(X_train1)
print('KNN Train Accuracy when K=3 : ',accuracy_score(y_train1,y_pred_train))

knn =KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train1,y_train1)
y_pred_5 = knn.predict(X_test1)
print("KNN Test Accuracy when K=5 :",accuracy_score(y_test1,y_pred_5))

knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(X_train1, y_train1)
y_pred_9 = knn.predict(X_test1)
print("KNN Test Accuracy when K=9 :",accuracy_score(y_test1,y_pred_9))


#Creating odd list to find best K
mylist = list(range(1,20))
neighbors = list(filter(lambda x : x % 2 != 0,mylist))
print(neighbors)

```

```python
ac_scores = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train1,y_train1)
    y_pred_val = knn.predict(X_test1)
    scores = accuracy_score(y_test1,y_pred_val)
    ac_scores.append(scores)
#Changing to misclassification error
MSE = [1- x for x in ac_scores]
best_k = neighbors[MSE.index(min(MSE))]
print('best K is : ',best_k)


plt.plot(neighbors,MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')

#cal accuracy score when k =13
knn_new = KNeighborsClassifier(n_neighbors = 13)
knn_new.fit(X_train1,y_train1)

#Predict test
y_pred_13_test = knn_new.predict(X_test1)
print('KNN Test Accuracy when K=13',accuracy_score(y_test1,y_pred_13_test))

#Predict train
y_pred_13_train = knn_new.predict(X_train1)
print('KNN Train Accuracy when K=13', accuracy_score(y_train1, y_pred_13_train))
```

#### NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
nav = GaussianNB()
nav.fit(X_train1,y_train1)

y_pred_nav = nav.predict(X_test1)
print('Naive Bayes Test Accuracy is :',accuracy_score(y_test1, y_pred_nav,normalize=True))

y_prev_nav_train = nav.predict(X_train1)
print('Naive Bayes Train Accuracy is : ',accuracy_score(y_train1,y_prev_nav_train))

print('Classification report \n')
print(classification_report(y_test1,y_pred_nav, labels=[1, 0]))
```

##### We can see our true positive numbers with value 1 has precision and recall  below 70%
